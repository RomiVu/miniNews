```python
def draw_bezier(json_file):
    with open(json_file) as f:
        data = json.load(f)
    
    l1 = data["line1"]
    l2 = data["line2"]
 
    p0 = data["p0"] # control  point
    p1 = data["p1"] # mid point of l1
    p2 = data["p2"] # mid point of l2

    features = data["features"]
    
    plt.plot(p0[0], p0[1], '*k')
    plt.plot(p1[0], p1[1], '*k')
    plt.plot(p2[0], p1[1], '*k')

    l1 = np.array(l1).T
    l2 = np.array(l2).T

    plt.plot(l1[0], l1[1])
    plt.plot(l2[0], l2[1])


    for i in range(len(features)):
        plt.plot(features[i][0], features[i][1], "*g")
    plt.show()
```

```python

    
def _handle_zeros_in_scale(scale, copy=False):
    if np.isscalar(scale):
        if scale == .0:
            scale = 1.
        return scale
    elif isinstance(scale, np.ndarray):
        scale[scale == 0.0] = 1.0
        return scale


def row_norms(X):
    norms = np.einsum('ij,ij->i', X, X)
    print(f"=== norms: {norms}")
    np.sqrt(norms, norms)
    print(f"=== np.sqrt(norms, norms) ===> norms: {norms}")
    return norms


def normalize(X, norm='l2', axis=0, copy=False, return_norm=True):
    X = X.T

    norms = row_norms(X)
    
    norms = _handle_zeros_in_scale(norms)
    
    print(f"##### before X /= norms[:, np.newaxis] : {norms}")
    X /= norms[:, np.newaxis]
    print(f"****** after X /= norms[:, np.newaxis] : {norms}")
    X = X.T

    return X, norms


class MyLinearRegression:
    def _preprocess_data(self, X, y):
#         print(f" before y shape : {y.shape} example: {y[0]}")
#         y = np.asarray(y, dtype=X.dtype)
#         print(f"after y shape : {y.shape} example: {y[0]}")

        X_offset = np.average(X, axis=0, weights=None)
        print(f"before X_offset: {X_offset}")
        X -= X_offset
        print(f"after X_offset,  X: {X}")
        
        X, X_scale = normalize(X)
        
        #print(f"after normalize,  X: {X}; X_scale : {X_scale}")
        
        y_offset = np.average(y, axis=0, weights=None)
        #print(f"before Y_offset: {y_offset}")
        y = y - y_offset
        #print(f"after y_offset,  Y: {y}")
        
        return X, y, X_offset, y_offset, X_scale
    
    def _set_intercept(self, X_offset, y_offset, X_scale):
        self.coef_ = self.coef_ / X_scale
        self.intercept_ = y_offset - np.dot(X_offset, self.coef_.T)

    def fit(self, X, y):
        X, y, X_offset, y_offset, X_scale = self._preprocess_data(X, y)
        
        print(f"X shape: {X.shape} ; Y shape: {y.shape} ; {len(X.shape)}; {type(y.shape)}")

        self.coef_, self._residues, self.rank_, self.singular_ = linalg.lstsq(X, y)
        self.coef_ = self.coef_.T
        print(f"self.coef_ : {self.coef_}")
        print(f"self._set_intercept(X_offset :{X_offset}, y_offset:{y_offset}, X_scale:{X_scale}) ")
        self._set_intercept(X_offset, y_offset, X_scale)
        return self
```

```python
def my_linear_regression(json_file):
    with open(json_file) as f:
        data = json.load(f)

    features = np.array(data["features"])
    labels = np.array(data["labels"])
    
    # print(f"features shape: {features.shape}; labels shape: {labels.shape}")
    
    lr = MyLinearRegression()
    lr.fit(features, labels)
    print(f"lr coffiecint :{lr.coef_}")
    print(f"lr intercept_ :{lr.intercept_}")
```
