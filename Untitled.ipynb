```python
from math import sqrt, pow

import matplotlib.pyplot as plt
import numpy as np

class Line:
    def __init__(self, *args, **kwargs):
        # todo list: validate
        if len(args) == 2:
            self.p1 = np.array(args[0])
            self.p2 = np.array(args[1])
            
        elif len(args) == 1 and isinstance(args[0], Line):
            self.p1 = Line.p1
            self.p2 = Line.p2
        else:
            raise ValueError("two many args or invaild args, initilization Line error.")
        self.length = 0
        self.slope = None
        self.normal_slope = None
        self.unit_vector = np.zeros(2)
        self.normal_unit_vector = np.zeros(2)
        self.set_params()
    
    def __str__(self):
        return f"Line:[{self.p1[0]},{self.p1[1]}, {self.p2[0]},{self.p2[1]}]"
    
    def __repr__(self):
        return f"Line:[{self.p1[0]},{self.p1[1]}, {self.p2[0]},{self.p2[1]}]"
    
    def set_params(self):
        self.set_length()
        if self.length == 0:
            raise ValueError("two points are the same one.")
            
        if self.p1[0] == self.p2[0]:
            self.slope = None
            self.unit_vector = np.array([0, 1])
        else:
            self.slope = (self.p1[1] - self.p2[1]) / (self.p1[0] - self.p2[0])
            self.unit_vector = np.array([(self.p1[0] - self.p2[0]) / self.length, (self.p1[1] - self.p2[1]) / self.length])

        if self.slope == None:
            self.normal_slope = 0
            self.normal_unit_vector = np.array([1, 0])
        elif self.slope == 0:
            self.normal_slope = None
            self.normal_unit_vector = np.array([0, 1])
        else:
            self.normal_slope = -1/self.slope
            angle = np.arctan(self.normal_slope)
            self.normal_unit_vector = np.array([np.cos(angle), np.sin(angle)])

    def set_length(self, method="educian"):
        self.length = sqrt(pow(self.p1[0] - self.p2[0], 2) + pow(self.p1[1] - self.p2[1], 2))

    def get_equation(self, form="normal"):
        if form == "normal":
        # ax + by + c = 0
            if self.slope == None:
                return 1, 0, -self.p1[0]
            return -self.slope, 1, self.slope*self.p1[0] - self.p1[1]
        elif form == "slope":
        # y = ax + b
            if self.slope == None:
                return None, None
            return self.slope, self.p1[1] - self.slope * self.p1[0]
        else:
            return None, None
        
    def get_mid_point(self):
        return (self.p1[0] + self.p2[0]) / 2, (self.p1[1] + self.p2[1]) / 2
    
    def line_position(self, line):
        # 0 parallel; 2 cross; 1 vertical
        a1, b1, c1 = self.get_equation()
        a2, b2, c2 = line.get_equation()
        if self.slope == line.slope:
            return 0, [None, None]
        else:
            t = a1 * b2 - a2 * b1
            if self.normal_slope == line.slope:
                return 1, [(-b2*c1+c2*b1)/t, (a2*c1-c2*a1)/t]
            return 2, [(-b2*c1+c2*b1)/t, (a2*c1-c2*a1)/t]
        
    def point_position(self, point):
        if self.slope == None:
            if point[0] > self.p1[0]:
                return 1 # rightside
            elif point[0] == self.p1[0]:
                return 0 # in the line
            else:
                return -1 # leftside
        else:
            a, b = self.get_equation(form="slope")
            t = a * point[0] + b
            if point[1] > t:
                return 1 # above the line
            elif t == point[1]:
                return 0 # in the line
            else:
                return -1 # below the line
            
    def point_boundary_position(self, point):
        ab = point - self.p1
        ac = self.p2 - self.p1
        
        d1 = ab.dot(ac)
        d2 = ac.dot(ac)
        
        t = self.point_position(point)
        
        if (0 < d1 < d2):
            return 1, t, np.cross(ac, ab)/np.linalg.norm(ac)
        elif (d1 == 0 or d1 == d2):
            return 0, t, 0
        else:
            return -1,t, np.cross(ac, ab)/np.linalg.norm(ac)
    
    def show(self):
        plt.plot([self.p1[0],self.p2[0]], [self.p1[1],self.p2[1]], '-g')
        plt.show()
        
    def show_vector(self):
        origin = [0], [0] # origin point
        plt.quiver(*origin, [self.unit_vector[0], self.normal_unit_vector[0]], [self.unit_vector[1], self.normal_unit_vector[1]], color=['r','b'], scale=10)
        plt.show()
        
    def test_points(self, num=1000):
        plt.plot([self.p1[0],self.p2[0]], [self.p1[1],self.p2[1]], '-g')

        for i in range(num):
            p = [np.random.randint(-10000, 10000), np.random.randint(-10000, 10000)]
            if self.point_position(p) == 1:
                color = "*b"
            elif self.point_position(p) == 0:
                print(p)
                color = "+r"
            else:
                color = "+k"
            
            plt.plot(p[0], p[1], color, label=f"p{i}")
        plt.show()
        
    def test_boundary_points(self, num=1000):
        plt.plot([self.p1[0],self.p2[0]], [self.p1[1],self.p2[1]], '-g')

        for i in range(num):
            p = np.array([np.random.randint(-1200, 1200), np.random.randint(-1200, 1200)])
            x, t, k = self.point_boundary_position(p)
            if x == 1:
                if t >= 0:
                    color = "*b"
                else:
                    color = "+b"
            elif x == 0:
                if t >= 0:
                    color = "*r"
                else:
                    color = "+r"
            else:
                if t >= 0:
                    color = "*k"
                else:
                    color = "+k"
            
            plt.plot(p[0], p[1], color, label=f"p{i}")
        plt.show()
        
    def test_lines(self, num=10):
        plt.plot([self.p1[0],self.p2[0]], [self.p1[1],self.p2[1]], '-g')

        for i in range(num):
            p1 = [np.random.randint(-10000, 10000), np.random.randint(-10000, 10000)]
            p2 = [np.random.randint(-10000, 10000), np.random.randint(-10000, 10000)]
            a = Line(p1, p2)
            
            x, p = self.line_position(a)

            if x == 0:
                color = "--b"
            elif x == 1:
                color = ":r"
            else:
                color = ":k"
            
            plt.plot([p1[0],p2[0]], [p1[1],p2[1]], color, label=f"a{i}")
            if p[0] != None:
                plt.plot(p[0],p[1], "*r", label=f"a{i}")
        plt.show() 
    
    def expand_parallelly(self, dis=0.0, out=True):
        pass
    
    def endpoint_expand_parallelly(self, dis=0.0, out=True):
        pass
    
    def shift_vertically(self, dis=0.0, up=True):
        if self.normal_unit_vector[1] >= 0:
            self.p1 = self.p1 + dis * self.normal_unit_vector
            self.p2 = self.p2 + dis * self.normal_unit_vector
        else:
            self.p1 = self.p1 - dis * self.normal_unit_vector
            self.p2 = self.p2 - dis * self.normal_unit_vector
    
    def test_shift_vertically(self):
        a = np.random.randint(0, 20)
        plt.plot([self.p1[0],self.p2[0]], [self.p1[1],self.p2[1]], '-g')
        print(f"dis= {a}")
        self.shift_vertically(dis=a)
        plt.plot([self.p1[0],self.p2[0]], [self.p1[1],self.p2[1]], '-k')
        
        plt.show()
    
def test_vertical():
    for i in range(100):
        p1 = [np.random.randint(-10000, 10000), np.random.randint(-10000, 10000)]
        p2 = [np.random.randint(-10000, 10000), np.random.randint(-10000, 10000)]
        a = Line(p1, p2)
        b = a.normal_unit_vector.dot(a.unit_vector)
        print(a.normal_unit_vector, a.unit_vector, b, np.isclose(b, 0.0))
```
